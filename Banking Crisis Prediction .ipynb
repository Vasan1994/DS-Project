# Encoding

```python
data = data.drop(['case', 'country'], axis=1)
data
```

```python
data.columns
```

```python
cc3_dummies = pd.get_dummies(data['cc3'])
data = pd.concat([data, cc3_dummies], axis=1)
data = data.drop('cc3', axis=1)
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
label_encoder=LabelEncoder()
data["banking_crisis"]=label_encoder.fit_transform(data["banking_crisis"])
Bank_mapping={index:label for index,label in enumerate(label_encoder.classes_)}
Bank_mapping
```

# Splitting

```python
from sklearn.model_selection import train_test_split
y = data['banking_crisis']
X = data.drop('banking_crisis', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

# Scaling

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
```

```python
X
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
```

# Logistic Regression

```python
 logistic_regression= LogisticRegression() 
logistic_regression.fit(X_train,y_train) 
```

```python
y_pred=logistic_regression.predict(X_test)
y_pred
```

```python
accuracy_score(y_test, y_pred) 
```

```python
con = confusion_matrix(y_test, y_pred)
con
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(con, annot=True, fmt='g', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
auc = roc_auc_score(y_test, y_pred)
auc
```
